```sql
SELECT 
  CAST(regexp_replace(price, '[$,]', '') AS DOUBLE) AS clean_price
FROM some_catalog.airbnb_data.berlin_listing
WHERE price IS NOT NULL 
  AND TRIM(price) <> ''
  AND regexp_replace(price, '[$,]', '') RLIKE '^[0-9]+$'
  AND CAST(regexp_replace(price, '[$,]', '') AS DOUBLE) BETWEEN 10 AND 1000





```
