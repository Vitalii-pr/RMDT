```python
%sql
select count(*) as num_of_orders From some_catalog.silver.orders
WHERE customer_id = 85909;
```

```python
%sql
SELECT 
    n.n_name AS nationality
FROM some_catalog.silver.orders o
JOIN some_catalog.silver.customer c ON o.customer_id = c.customer_id
JOIN some_catalog.silver.nation n ON c.nation_id = n.n_nationkey
GROUP BY n.n_name, c.customer_id
ORDER BY COUNT(o.order_id) DESC
LIMIT 1;
```
